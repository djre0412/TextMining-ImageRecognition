# **Problema 1**

## *Función*

```python
def cambiocolor(carpeta,imagen,formatoimagen,canal):

    img=cv2.imread(carpeta+"/"+ imagen+formatoimagen)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    nameimg=""
    alto=img.shape[0]
    ancho=img.shape[1]

    img_salida=np.zeros_like(img)


    for i in range(0,alto):
        for j in range(0,ancho):
            pixel=img[i,j]

            
            red=pixel[0]
            green=pixel[1]
            blue=pixel[2]

            
            if(canal==1):
                img_salida[i,j]=(0,0,blue)
                nameimg="Blue"


            elif(canal==2):
                img_salida[i,j]=(0,green,0)
                nameimg="green"

            elif(canal==3):
                img_salida[i,j]=(red,0,0)
                nameimg="red"

            elif(canal==10):
                img_salida[i,j]=(0,green,blue)
                nameimg="verde_azul"
            elif(canal==20):
                img_salida[i,j]=(red,green,0)
                nameimg="rojo_verde"
            elif(canal==30):
                img_salida[i,j]=(red,0,blue)
                nameimg="rojo_azul"
                
    plt.imshow(img_salida)
    plt.show()
    img_salida=cv2.cvtColor(img_salida,cv2.COLOR_RGB2BGR)  
    cv2.imwrite(carpeta+"/"+imagen+nameimg+".jpg", img_salida)
```

## *Imagen Monocromatica*

```python
cambiocolor("Paris2024","Paris2024",".jpg",3)
```

# **Problema #2**

## *Función*

```python
def tresgrises_a_color(carpeta,formatoimagen,imagenGrisAzul, imagenGrisRojo, imagenGrisVerde):

    imgazul_gray2=cv2.cvtColor(cv2.imread(carpeta+"/"+imagenGrisAzul+formatoimagen),cv2.COLOR_BGR2RGB)
    imgrojo_gray2=cv2.cvtColor(cv2.imread(carpeta+"/"+imagenGrisRojo+formatoimagen),cv2.COLOR_BGR2RGB)
    imgverde_gray2=cv2.cvtColor(cv2.imread(carpeta+"/"+imagenGrisVerde+formatoimagen),cv2.COLOR_BGR2RGB)
    
    alto=imgazul_gray2.shape[0]
    ancho=imgazul_gray2.shape[1]

    img_salida2=np.zeros_like(imgazul_gray2)

    for i in range(0,alto):
        for j in range(0,ancho):
            pixelazul=imgazul_gray2[i,j]
            pixelrojo=imgrojo_gray2[i,j]
            pixelverde=imgverde_gray2[i,j]

                    
            red=pixelrojo[0]
            green=pixelverde[0]
            blue=pixelazul[0]

            img_salida2[i,j]=(red,green,blue)
    plt.imshow(img_salida2)
    plt.show()
    img_salida2=cv2.cvtColor(img_salida2,cv2.COLOR_RGB2BGR)  
    cv2.imwrite(carpeta+"/imagen a color.jpg", img_salida2)
```

# **Problema 4**

## *Función*

```python
def histogramas_canal(carpeta,imagen,formatoimagen):

    img=cv2.imread(carpeta+"/"+ imagen+formatoimagen)

    alto=img.shape[0]
    ancho=img.shape[1]

    red=0
    green=0
    blue=0
    grey=0

    img_salida_grey=np.zeros((alto,ancho,1))

    for i in range(0,alto):
        for j in range(0,ancho):
            pixel=img[i,j]
            img_salida_grey[i,j]=pixel.mean()

            red=red+pixel[2].astype(np.int64)
            green=green+pixel[1].astype(np.int64)
            blue=blue+pixel[0].astype(np.int64)
            grey=grey+pixel[0].astype(np.int64)

            

    img_salida_grey=img_salida_grey.astype(np.uint8)

    hist1=cv2.calcHist([img],[2],None,[256],[0,256])
    hist2=cv2.calcHist([img],[1],None,[256],[0,256])
    hist3=cv2.calcHist([img],[0],None,[256],[0,256])
    hist4=cv2.calcHist([img_salida_grey],[0],None,[256],[0,256])


    mean_his0 = red/(alto*ancho)
    mean_his1 = green/(alto*ancho)
    mean_his2 = blue/(alto*ancho)
    mean_his3 = grey/(alto*ancho)


    plt.figure(figsize=(6,3))
    plt.plot(hist1,color="Red",label="Rojo")
    plt.axvline(x=mean_his0, color='Red', linestyle='--', label='Media Rojo')

    plt.xlabel("Intensidad del pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma pixeles por canal")
    plt.legend()
    plt.show()


    plt.figure(figsize=(6,3))
    plt.plot(hist2,color="green",label="Verde")
    plt.axvline(x=mean_his1, color='green', linestyle='--', label='Media Verde')

    plt.xlabel("Intensidad del pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma pixeles por canal")
    plt.legend()
    plt.show()

    plt.figure(figsize=(6,3))
    plt.plot(hist3,color="Blue",label="Azul")
    plt.axvline(x=mean_his2, color='blue', linestyle='--', label='Media Azul')
    plt.xlabel("Intensidad del pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma pixeles por canal")
    plt.legend()
    plt.show()


    plt.figure(figsize=(6,3))
    plt.plot(hist4,color="grey",label="Escala grises")
    plt.axvline(x=mean_his3, color='grey', linestyle='--', label='Media grises')
    plt.xlabel("Intensidad del pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma pixeles Escala de Grises")
    plt.legend()
    plt.show()
```

## *Histogramas*

```python
histogramas_canal('Simpsons','TheSimpsons','.jpg')
```

# **Problema #5**

## *Escala de grises ponderada*

Es una técnica utilizada en el procesamiento de imágenes para convertir una imagen a escala de grises, pero asignando diferentes pesos a los canales de color (rojo, verde y azul) para obtener resultados más precisos o adaptados a una tarea específica.

Al asignar pesos diferentes a cada canal, se puede crear una representación en escala de grises que se asemeje más a lo que el ojo humano percibiría.



La fórmula general para calcular el valor de un píxel en escala de grises ponderada es:

***Gris = 0.30R + 0.60G + 0.10B***

## *Función*

```python
def grises_ponderado(carpeta,imagen,formatoimagen):

    img=cv2.imread(carpeta+"/"+ imagen+formatoimagen)




    alto=img.shape[0]
    ancho=img.shape[1]

    img_salida_grey=np.zeros((alto,ancho,1))

    for i in range(0,alto):
        for j in range(0,ancho):
            pixel=img[i,j]

            red=pixel[2]
            green=pixel[1]
            blue=pixel[0]

            img_salida_grey[i,j]=(red*0.30)+(green*0.60)+(blue*0.10)

    plt.imshow(img_salida_grey, cmap="gray")
    plt.show()

    
    cv2.imwrite(carpeta+"/"+imagen+"_salida_gray_ponderado"+".jpg", img_salida_grey)
```

## *Imagen en escala de grises ponderada*

```python
grises_ponderado('Simpsons','TheSimpsons','.jpg')
```

## *Espacio de color HSV*

El espacio de color HSV (Hue, Saturation, Value, o Matiz, Saturación, Valor en español) es un modelo de representación de color que se utiliza en muchas aplicaciones de procesamiento de imágenes, especialmente en aquellas relacionadas con la percepción visual humana. A diferencia del espacio RGB (Rojo, Verde, Azul), que se basa en la mezcla de luces primarias, el HSV se enfoca en las características intuitivas del color que los humanos percibimos.

* **Matiz (Hue):** Corresponde al tono o color puro, como el rojo, verde o azul. Se representa como un ángulo en un círculo de colores, donde 0 grados suele corresponder al rojo.
* **Saturación (Saturation):** Indica la intensidad o pureza del color. Un valor de saturación alto significa un color vibrante y puro, mientras que un valor bajo indica un color más apagado o grisáceo.
* **Valor (Value):** Representa la luminosidad o brillo del color. Un valor alto indica un color claro, mientras que un valor bajo indica un color oscuro.
